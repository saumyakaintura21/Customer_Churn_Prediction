load libraries and data 

```python
import pandas as pd
```

```python
file_name = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'  
df = pd.read_csv(file_name)
```

```python
print("Data loaded into a DataFrame named 'df'.")
df.info()
```

```python
# Convert 'TotalCharges' to numeric. 'errors='coerce'' turns non-numeric values (blanks) into NaN.
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Print the number of missing values created (should be small, around 11)
print("Missing values in TotalCharges:", df['TotalCharges'].isnull().sum())
```

```python
# Drop rows with any missing values (only affects the 11 rows in TotalCharges)
df.dropna(inplace=True) 

# Re-run info to confirm the missing values are gone and TotalCharges is now float64
df.info()
print("\nNew DataFrame shape:", df.shape)
```
